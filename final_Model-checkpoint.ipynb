## Cleaning the dataset

```python
# Dropping the 72591 row of the dataset
D.drop(72591,inplace=True)
```

```python
D.loc[72585:72594,["speedX","speedY","speedY"]]
```

```python
df = D.replace(['[]'],'0')
D['angle'] = D['angle'].str.strip('[]')
D['curLapTime'] = D['curLapTime'].str.strip('[]')
D['distFromStart'] = D['distFromStart'].str.strip('[]')
D['distRaced'] = D['distRaced'].str.strip('[]')
D['gear'] = D['gear'].str.strip('[]')
D['lastLapTime'] = D['lastLapTime'].str.strip('[]')
D['racePos'] = D['racePos'].str.strip('[]')
D['rpm'] = D['rpm'].str.strip('[]')
D['speedX'] = D['speedX'].str.strip('[]')
D['speedY'] = D['speedY'].str.strip('[]')
D['speedZ'] = D['speedZ'].str.strip('[]')
D['trackPos'] = D['trackPos'].str.strip('[]')
D['accel'] = D['accel'].str.strip('[]')
D['brake'] = D['brake'].str.strip('[]')
D['steer'] = D['steer'].str.strip('[]')
```

## Some rows contains noise or irregular value so replacing it

```python
# The row 72832 contains wrong data so cleaning it
D.loc[72832,"angle"]=D.loc[72831,"angle"]
D.loc[72832,"trackpos"]=D.loc[72831,"trackpos"]
D.loc[72832,"steer"]=D.loc[72831,"steer"]
D.loc[72832,"speedX"]=D.loc[72831,"speedX"]
D.loc[72832,"speedY"]=D.loc[72831,"speedY"]
D.loc[72832,"speedZ"]=D.loc[72831,"speedZ"]
```

## Converting the object to int to give it to the model

```python
# Converting the attribute values to the numeric values
D['angle'] = pd.to_numeric(D['angle'])
D['trackpos'] = pd.to_numeric(D['trackpos'])
D['steer'] = pd.to_numeric(D['steer'])
D['speedX'] = pd.to_numeric(D['speedX'],errors="coerce")
D['speedY'] = pd.to_numeric(D['speedY'],errors="coerce")
D['speedZ'] = pd.to_numeric(D['speedZ'],errors="coerce")

```

## Dropping the null rows 

```python
D.dropna(inplace=True)
```

## Now applying the model on the data

## Importing the test train and split

```python
from sklearn.model_selection import train_test_split
```

```python
X=D[["angle","trackpos","speedX","speedY","speedZ"]]
y=D["steer"]
```

## 80% testing and 20% training

```python
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
# Length of the training set
len(X_train)
```

```python
# Making linear regression object
model=linear_model.LinearRegression()
```

```python
# Fitting the model
model.fit(X_train,y_train)
```

```python
model.predict(X_test)
```

```python
# Calculating the accuracy
model.score(X_test,y_test)
```

```python
X1=D[["speedX","speedY","speedZ"]]
y1=D["brake"]
X10_train, X10_test, y10_train, y10_test=train_test_split(X1,y1,test_size=0.2,random_state=10)
model2=linear_model.LinearRegression()
model2.fit(X10_train,y10_train)
model2.predict(X10_test)
```
